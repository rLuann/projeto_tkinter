# Projeto de Programação A - 2026-1

Neste projeto será elaborado um programa para construir desenhos no estilo das apps Google Drawings e Libre Office Draws.

O projeto deverá utilizar o Tkinter, a biblioteca padrão de Python para construção de GUIs.

Será disponibilizada uma implementação inicial simples e incompleta, estruturada imperativamente, cujo link é fornecido abaixo:

https://drive.google.com/drive/folders/1pKYGBpra9JAw9yuGuXvUhIs341lDnjYD?usp=sharing

O projeto deverá ser desenvolvido em equipes de dois ou três alunos (preferencialmente três).
As equipes deverão:
  - Refatorar o projeto para uma estrutura de programa Orientada a Objetos.
  - Adicionar funcionalidades.
  - Aplicar alguns padrões arquiteturas e de projeto.
  - Desenvolver o sistema utilizando as ferramentas Git e Github para trabalho colaborativo.

As entregas devem ser feitas criando tags de versionamento do projeto no Git/Github.
  + Algumas entregas exigirão o registro de commits individuais no Github.
  + Atrasos nas entregas serão penalizados. Cada dia de atraso tirará 0.5 pontos.

As tarefas das entrega devem ser divididas entre os integrantes do grupo. Cada aluno deverá registrar seus commits individuais no Github, que serão utilizados como parâmetro para definição das notas.

## Entrega 1 (11/07 - tag: imperativa.1)
Leitura de colab sobre Git e Github e criação do projeto no Github.

Adicionar funcionalidades que permitam:
  + Desenhar retângulos.
  + Desenhar ovais.
  + Desenhar círculos.
  + Que os desenhos individuais possam ter bordas com cor.
  + Que os desenhos individuais tenham cor de preenchimento.


## Entrega 2 (13/07 tag: OO.1)
Refatorar o sistema para seguir uma abordagem OO.
  + Definir uma hierarquia de classes (`Figura`) para contemplar os diversos tipos de desenhos: retângulos, mão livre, etc.
  + Adequar o programa para utilizar a hierarquia de figuras (`Figura`).
  + Adicionar desenhos de polígonos.
  + Separar o código em módulos (Classe `Figura` e subclasses em um arquivo separado, por exemplo)
    
## Entrega 3 (16/07 tag: OO.MVC.1)
Refatorar para utilizar o padrão MVC.
  + Definir as classes do modelo (Figuras, Desenho, ...)
  + Definir uma classe ou classes para a visão
  + Definir uma classe ou classes para o(s) controlador(es)

Recomendação: usar a seguinte estrutura de arquivos e pastas
    
    nome-do-projeto/
    ├── .git/                     # Controlado pelo Git
    ├── .gitignore                # Arquivos e pastas ignorados pelo Git
    ├── src/                      # Código-fonte principal do projeto
    │   └── nome_do_projeto/      # Pacote python  
    │       ├── main.py           # O programa principal
    │       ├── modelo            # Classes do Model
    │       │   └── ...           #
    │       ├── visao             # Classes da View
    │       │   └── ...
    │       └── controlador       # Classes Controllers
    │           └── ...


## Entrega 4 (19/07 tag: OO.State.1)
  + Utilizar o Padrão State para evitar código condicional nos métodos do(s) controlador(es)
  + Adicionar as funcionalidades Salvar e Abrir
